```python
import pandas as pd
```

```python
df = pd.read_csv('CSE All-Share Historical Data.csv')
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
cse_df = pd.DataFrame()
```

```python
price_list = [float(item.replace(',', '')) for item in df['Price'].tolist()]
```

```python
cse_df['date'] = df['Date']
cse_df['Price'] = price_list
```

```python
cse_df.index = pd.to_datetime(cse_df.index)
```

```python
cse_df.drop(['date'], axis=1, inplace=True)
cse_df.index = df['Date']
```

```python
cse_df
```

```python
cse_df.index = pd.to_datetime(cse_df.index)
```

```python
f, ax = plt.subplots(1, 1)
ax.plot(cse_df['Price'])
ax.set_title('Time-Series graph all share price index over an year')
ax.tick_params(axis='x', rotation=45)
plt.show()
plt.close()
```

```python
cse_df.sort_index(ascending=True, inplace=True)
```

```python
rollingmean = cse_df['Price'].rolling(window=200, center=False).mean()
rollingstd = cse_df['Price'].rolling(window=200, center=False).std()
#orig = plt.plot(cse_df['Price'], 
#                color='blue',
#                label='Original')

ewmmean = plt.plot(rollingmean, 
                color='blue',
                label='Rolling Mean')
std = plt.plot(rollingstd,
               color='black',
               label='Rolling Std')
log = plt.plot(cse_df['Price'],
               color='red',
               label='Original')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation for all share price')
plt.xticks(rotation=45)
plt.show(block=False)
plt.close()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(cse_df['log_price'], autolag='AIC')
```

```python
import numpy as np
```

```python
cse_df['log_price'] = cse_df['Price'].apply(lambda x: np.log(x))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(cse_df['Price'], freq = 365)
```

```python
cse_df.loc[:,'trend'] = decomposition.trend
cse_df.loc[:, 'seasonal'] = decomposition.seasonal
cse_df.loc[:, 'residual'] = decomposition.resid
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (15, 5), sharex = True)
ax1.plot(cse_df['trend'], label='Original')
ax1.legend(loc='best')
ax1.tick_params(axis = 'x', rotation = 45)

ax2.plot(cse_df['seasonal'], label='Seasonal')
ax2.legend(loc='best')
ax2.tick_params(axis = 'x', rotation = 45)

ax3.plot(cse_df['residual'], label='Residual')
ax3.legend(loc='best')
ax3.tick_params(axis = 'x', rotation = 45)
```

```python
new_df = cse_df.dropna()
adfuller(new_df['residual'], autolag='AIC')
```
